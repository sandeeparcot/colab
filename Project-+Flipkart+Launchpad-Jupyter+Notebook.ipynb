- ### 3.5: Display the names of the top 5 categories based on total views, with minimum 10,000 comments on each video. (4 marks)

```python
# Write your code here to select top 5 category ids with minimum 10000 comments in each video (2 marks)

#performing a pivot table by category_id and getting the total number of views by category and assigning into a new dataframe res1
res1 = pd.pivot_table(videos,index='category_id',aggfunc=np.sum)
#filtering new dataframe by comments minimum of 10000 and selecing the top 5 categories and assigning to ids list using tolist() function
ids = res1[res1['comment_count']>=10000].sort_values('views',ascending = 0)[:5].index[0:5].tolist()
ids

```

```python
#importing the categories.csv using read_csv and we are using the id column as an index rather than the default index
Categories=pd.read_csv("Categories.csv",index_col="id")

```

```python
#create a function that takes in a list of ids and outputs a list of category names coresponding to those ids (2 marks)
def id_to_name(id_list):
    
    # write your code here
    clist=[] #assigning the output list variable
    for i in id_list:
        clist.append(Categories.loc[i,'Category'])  #appending the values of category name by using loc function to fectch the name based on the index category id
    return clist
    

print(id_to_name(ids))
```


- ### 4.2: For the two selected channels, check which one has a higher number of negative responses per positive response. To do this, follow these steps:
    - First, create two new data frames for each channel and sort the rows by the number of views. (2 marks)
    - In each data frame, create a new column containing the ratio of dislikes/likes. (2 marks)
    - Finally, check the summary statistics of both these (new) columns. (2 marks)
    - Compare these results and identify which channel publishes more controversial videos (higher dislikes/likes value). (2 marks)





```python
#Create and sort channel 1
c1 = videos[videos['channel_title']=='PewDiePie'].sort_values('views',ascending = 0)
c1.shape
```

```python
#Create and sort channel 2
c2 = videos[videos['channel_title']=='T-Series'].sort_values('views',ascending = 0)
c2.shape
```

```python
#Create 'Ratio' column for channel 1
# creating a new column ratio and assigning ration by dividing the dislike comments by like comments and getting %
c1['Ratio'] = c1.apply(lambda i: round(100*(i.dislikes/i.likes), 2),axis=1)
c1.describe().Ratio # getting the summary statistics based on the newly created ration column
```

```python
#Create 'Ratio' column for channel 2
# creating a new column ratio and assigning ration by dividing the dislike comments by like comments and getting %
c2['Ratio'] = c2.apply(lambda i: round(100*(i.dislikes/i.likes), 2),axis=1)  
c2.describe().Ratio  # getting the summary statistics based on the newly created ration column
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Check summary statistics of both channel's ratios 
plt.boxplot(c1.Ratio,vert=False) # plotting a box plot to understand more on the outliers and median values
plt.show()
```

```python
plt.boxplot(c2.Ratio,vert=False)
plt.show()
```
